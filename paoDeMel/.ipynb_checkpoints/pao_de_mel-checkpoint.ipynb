# Limpeza de Dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando os arquivos CSV

compras = pd.read_csv("base_compras.csv")
departamentos = pd.read_csv("base_departamentos.csv")
produtos = pd.read_csv("base_produtos.csv")
tipo_produto = pd.read_csv("base_tipo_de_produto.csv")
```

```python
# Explorando as primeiras linhas de cada dataset

# Tabela Compras

compras.head()
```

```python
# Tabela Departamentos

departamentos.head()
```

```python
# Tabela Produtos

produtos.head()
```

```python
# Tabela Tipo de Produtos

tipo_produto.head()
```

```python
# Verificando valores ausentes

def check_missing_values(df, name):
    missing = df.isnull().sum()
    print(f"\nValores ausentes em {name}:")
    print(missing[missing > 0])

check_missing_values(compras, "Compras")
check_missing_values(departamentos, "Departamentos")
check_missing_values(produtos, "Produtos")
check_missing_values(tipo_produto, "Tipo de Produto")
```

```python
# Verificando tipos de dados

def check_data_types(df, name):
    print(f"\nTipos de dados em {name}:")
    print(df.dtypes)
    
check_data_types(compras, "Compras")
check_data_types(departamentos, "Departamentos")
check_data_types(produtos, "Produtos")
check_data_types(tipo_produto, "Tipo de Produto")
```

```python
# Removendo duplicatas

def remove_duplicates(df, name):
    before = df.shape[0]
    df.drop_duplicates(inplace=True)
    after = df.shape[0]
    print(f"\nRemovidas {before - after} duplicatas de {name}.")
    return df

compras = remove_duplicates(compras, "Compras")
departamentos = remove_duplicates(departamentos, "Departamentos")
produtos = remove_duplicates(produtos, "Produtos")
tipo_produto = remove_duplicates(tipo_produto, "Tipo de Produto")
```

```python
# Padronizando colunas (Exemplo: nomes em minúsculo e sem espaços)

def standardize_columns(df):
    df.columns = df.columns.str.lower().str.replace(" ", "_")
    return df

compras = standardize_columns(compras)
departamentos = standardize_columns(departamentos)
produtos = standardize_columns(produtos)
tipo_produto = standardize_columns(tipo_produto)
```

```python
# Criando dataset final com merge das tabelas

dataset_final = compras.merge(produtos, on='product_id', how='left')
dataset_final = dataset_final.merge(departamentos, on='department_id', how='left')
dataset_final = dataset_final.merge(tipo_produto, on='aisle_id', how='left')

dataset_final
```

```python
# Salvando dataset final limpo

dataset_final.to_csv("dataset_limpo.csv", index=False)
print("\nDataset final salvo como 'dataset_limpo.csv'")
```

```python
# Criando um arquivo .gitignore para evitar versionamento do dataset grande

with open(".gitignore", "w") as gitignore:
    gitignore.write("dataset_limpo.csv\n")
print("Arquivo .gitignore criado para ignorar 'dataset_limpo.csv'")
```

# Analise Exploratória dos Dados

```python
# Lendo os dados do arquivo CSV

df = pd.read_csv('dataset_limpo.csv')

df.info()
```

```python
df.head()
```

```python
# Contagem de valores por Departamento

department_count = df['department'].value_counts()

department_count
```

```python
# Relação entre departamento e número de compras

plt.figure(figsize=(12, 6))
sns.countplot(x='department', data=df, order=df['department'].value_counts().index)
plt.title("Número de Compras por Departamento")
plt.xlabel("Departamento")
plt.ylabel("Contagem")
plt.xticks(rotation=45)
plt.show()
```

```python
# Contagem de valores por Dia da Semana

dow_count = df['order_dow'].value_counts()

dow_count
```

```python
# Distribuição das Compras ao Longo da Semana

if 'order_dow' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(x='order_dow', data=df, order=sorted(df['order_dow'].unique()))
    plt.title("Número de Compras por Dia da Semana")
    plt.xlabel("Dia da Semana")
    plt.ylabel("Contagem")
    plt.show()
```

```python
# Contagem de valores por Hora do Dia

hour_count = df['order_hour_of_day'].value_counts()

hour_count
```

```python
# Relação entre Compras e Hora do dia

plt.figure(figsize=(12, 6))
sns.countplot(x='order_hour_of_day', data=df, order=sorted(df['order_hour_of_day'].unique()))
plt.title("Número de Compras por Hora do Dia")
plt.xlabel("Hora do Dia")
plt.ylabel("Contagem")
plt.xticks(rotation=45)
plt.show()
```

```python
# Produtos Mais Vendidos

plt.figure(figsize=(12, 6))
df['product_name'].value_counts().head(5).plot(kind='bar')
plt.title("Top 5 Produtos Mais Comprados")
plt.xlabel("Nome do Produto")
plt.ylabel("Número de Compras")
plt.show()
```

```python
# Frequência de Compras por Cliente

if 'user_id' in df.columns:
    plt.figure(figsize=(12, 6))
    df['user_id'].value_counts().head(5).plot(kind='bar')
    plt.title("Top 5 Clientes com Mais Compras")
    plt.xlabel("ID do Cliente")
    plt.ylabel("Número de Compras")
    plt.show()
```

```python
# Contagem de valores por Tipo de Produto

aisle_count = df['aisle_id'].value_counts()

aisle_count
```

```python
# Relação entre Compras e Tipo de Produto

plt.figure(figsize=(12, 30))
sns.countplot(y='aisle_id', data=df, order=sorted(df['aisle_id'].unique()))
plt.title("Número de Compras por Tipo de Produto")
plt.xlabel("Contagem")
plt.ylabel("Tipo de Produto")
plt.xticks(rotation=45)
plt.show()
```

```python
# Concentração de Tipo de Produto por Hora do Dia

if 'order_hour_of_day' in df.columns and 'aisle_id' in df.columns:
    plt.figure(figsize=(14, 25))
    sns.heatmap(df.pivot_table(index='aisle_id', columns='order_hour_of_day', aggfunc='size', fill_value=0), cmap="coolwarm")
    plt.title("Concentração de Tipo de Produto por Hora do Dia")
    plt.xlabel("Hora do Dia")
    plt.ylabel("ID do Tipo de Produto")
    plt.show()
```
